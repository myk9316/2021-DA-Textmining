# 2. 데이터 전처리

```python
# 각각의 기사를 tab 단위로 분리 (제목, 날짜, 신문사, 내용)
new_data = []
for d in data:
    new_data.append(d.strip().split('\t'))
new_data[0]
```

```python
# 타이틀만 따로 추출후 먹기리와 관련된 내용이 아닌 것들은 본문을 확인하였음. 확인 후 del_title에 관련 없어보이는 단어들을 넣었음
title = []
for i in range(len(new_data)):
    title.append(new_data[i][0])
title
```

```python
del_title = ['부고', '국감', '알림', '인물 동정', '인사']
```

```python
preprocessed_data = []

for article in new_data:
    if len(article) != 4: #제목, 날짜, 신문사, 본문 4개가 전부 없는 데이터는 전처리 데이터에 포함 안함
        continue
    elif int(article[1][:4]) not in range(2012, 2018): # 2012년~2017년 사이에 없는 데이터는 포함 안함
        continue
    elif any(word in article[0] for word in del_title): # 제목에 del_title에 있는 단어가 들어가는 경우 포함 안함
        continue
    else:
        preprocessed_data.append(article)
len(preprocessed_data)
```

```python
#  무단전재 및 재배포 금지는 분석에 필요없는 단어라고 생각되서 삭제함
for i in range(len(preprocessed_data)):
    if '무단전재 및 재배포 금지' in preprocessed_data[i][-1]:
        preprocessed_data[i][-1] = preprocessed_data[i][-1].replace('무단전재 및 재배포 금지', '')
    elif '무단 전재 및 재배포 금지' in preprocessed_data[i][-1]:
        preprocessed_data[i][-1] = preprocessed_data[i][-1].replace('무단 전재 및 재배포 금지', '')
len(preprocessed_data)
```

```python
# 제목과 본문 합침
title_concat = [article[0] + ' ' + article[-1] for article in preprocessed_data]
title_concat[:2]
```

```python
data_df = pd.DataFrame(title_concat)
data_df.to_csv('preprocessing_data.txt', index=False, header=False)
```
